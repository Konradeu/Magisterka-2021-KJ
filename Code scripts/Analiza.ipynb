```python
import numpy as np
import pickle
import glob
from scipy.stats import mannwhitneyu
```

```python
NumOfBatches = [4,8,12,16]
ImageType = ['Cutted_pictures','Normal','Full_pictures']
NumOfEpochs = [500,1000,1500,3000,5000,10000]
```

```python
# Wyniki dla dekodera
dataDir = './decoder_results/'

resultsDecoder = []
for num in NumOfBatches:
    for typ in ImageType:
        for epochs in NumOfEpochs:
            filename = dataDir + str(num) + ' batches_' + typ + '__' + str(epochs)+ ' epochs_dict.pkl'
            with open(filename, 'rb') as f:
                x = pickle.load(f)
            real = []
            fake = []
            for sample in x:
                if sample['label'] == 1:
                    real.append(sample['score'])
                else:
                    fake.append(sample['score'])
                    
            #Statistical test        
            U1, p = mannwhitneyu(real, fake)
            resultsDecoder.append(p)
            print(num,typ,epochs,float('%.3g' % p))
            
```

```python
#Wyniki dla OneClass
dataDir = './oneclass_results/'

resultsOneClass = []
experiments = []
for num in NumOfBatches:
    for typ in ImageType:
        for epochs in NumOfEpochs:
            #Fake
            filename = dataDir + 'F_OneClass/' + str(num) + ' batches_' + typ + '_' + str(epochs)+ ' epochs_score_gen.pkl'
            with open(filename, 'rb') as f:
                fake = pickle.load(f)
                
            #Real
            filename = dataDir + 'R_OneClass/' + str(num) + ' batches_' + typ + '_score_test.pkl'
            with open(filename, 'rb') as f:
                real = pickle.load(f)
             
            #Statistical test
            U1, p = mannwhitneyu(real, fake)
            resultsOneClass.append(p)
            experiments.append((num,typ,epochs))
            print(num,typ,epochs,float('%.3g' % p))
```

```python

sigLevel = 0.001 #0.001 #0.01 #0.05

sumFlag1 = 0
sumFlag2 = 0
for e,o,d in zip(experiments,resultsOneClass,resultsDecoder):
    flag1 = 1 if o<sigLevel else 0
    flag2 = 1 if d<sigLevel else 0

    sumFlag1 += flag1
    sumFlag2 += flag2
    
    print(e,flag1,flag2)

print('in total ',sumFlag1,sumFlag2)
```
